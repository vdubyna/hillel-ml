13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
def multiply_or_divide(column):
    def transform(val, avg):
        if val > avg:
            return val * 2
        else:
            return val / 4

    vfunc = np.vectorize(transform)

    return vfunc(column, column.mean())
```

14. Застосувати отриману функцію до 3-ї колонки

```python
updated_col = multiply_or_divide(iris[:, 2])
updated_col
```
